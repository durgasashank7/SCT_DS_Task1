```python
import matplotlib.pyplot as plt
import pandas as pd

# Population Dataset
data = {
    'Age Group': ['0-18', '19-35', '36-50', '51-65', '66+'],
    'Population': [12000, 19000, 15000, 9000, 4000]
}

df = pd.DataFrame(data)

# Dark futuristic theme
plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(10, 6))

# Neon colors
colors = ['#00F5FF', '#00FF9F', '#FFD700', '#FF6EC7', '#FF3131']

bars = ax.bar(
    df['Age Group'],
    df['Population'],
    color=colors,
    edgecolor='white',
    linewidth=2
)

# Title and labels
ax.set_title(
    'Population Distribution by Age Group',
    fontsize=20,
    color='cyan',
    pad=20,
    fontweight='bold'
)

ax.set_xlabel(
    'Age Groups',
    fontsize=14,
    color='white',
    fontweight='bold'
)

ax.set_ylabel(
    'Population Count',
    fontsize=14,
    color='white',
    fontweight='bold'
)

# Grid effect
ax.grid(
    color='gray',
    linestyle='--',
    linewidth=0.5,
    alpha=0.3
)

# Value labels
for bar in bars:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        yval + 300,
        yval,
        ha='center',
        fontsize=11,
        color='white',
        fontweight='bold'
    )

# Background
fig.patch.set_facecolor('#0D1117')
ax.set_facecolor('#161B22')

plt.show()
```
